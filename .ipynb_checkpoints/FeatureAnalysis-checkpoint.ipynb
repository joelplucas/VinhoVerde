```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import preprocessing

import warnings
warnings.filterwarnings('ignore')
```

```python
#import data
train = pd.read_csv('input/train.csv')
test = pd.read_csv('input/test.csv')
```

```python
def print_PCF(data):
    colormap = plt.cm.RdBu
    plt.figure(figsize=(14,12))
    plt.title('Pearson Correlation of Features', y=1.05, size=15)
    data['Age'] = data['Age'].astype(float)
    data['Fare'] = data['Fare'].astype(float)
    data['Cabin'] = data['Cabin'].astype(float)
    data['Title'] = data['Title'].astype(float)
    sns.heatmap(data.corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
def to_integer(data, feature):
    le = preprocessing.LabelEncoder()
    le.fit(data[feature])
    data[feature] = le.transform(data[feature]) 
    
def ferature_eng(data):
    new_data = data.copy()
    new_data['bin_Age'] = pd.cut(x=new_data['Age'].dropna(), right=True, bins=(0,3,12,60,100), labels=[0,1,2,3]).astype(float)
    new_data['bin_Fare'] = pd.cut(x=new_data['Fare'].dropna(), right=True, bins=(0,7,10,50,80,600), labels=[0,1,2,3,4]).astype(float)
    new_data['Sex'] = new_data['Sex'].map( {'female': 0, 'male': 1}).astype(float)
    new_data['Embarked'] = new_data['Embarked'].map( {'S': 1, 'C': 2, 'Q': 3} ).astype(float)
    new_data['Family_Members'] = (data['SibSp'] + data['Parch']).astype(float)
    new_data['bin_Family_Members'] = pd.cut(x=new_data['Family_Members'], right=False, bins=(0,1,3,4,6,7,999), labels=[0,1,2,3,4,5]).astype(float)
    new_data['IsAlone'] = (new_data['Family_Members']>0).astype(float)
    new_data['Cabin'] = new_data['Cabin'].fillna('X').str[0:1]
    to_integer(new_data, 'Cabin')
    new_data['Cabin'] = new_data['Cabin'].map({1:0,3:0,4:0,7:0,2:1,5:1,0:2,6:2,8:3})
    new_data['Title'] = new_data['Name'].apply(lambda x : x.split(",")[1].split(".")[0].strip())
    to_integer(new_data, 'Title')
    new_data['Title'] = new_data['Title'].map({1:0,3:0,6:0,7:0,5:1,9:1,10:1,13:1,15:1,16:1,2:2,4:2,14:2,8:3,12:3,11:4})
    return new_data
```

```python
#find Collinearities + features with high information gain (high correlation with the Label('Survived'))
print_PCF(ferature_eng(train).drop(['PassengerId'],axis=1))
```

```python
def compare_integer_to_label(data, feature):
    g = sns.factorplot(x=feature, y="Survived", data=data, kind="bar", size=4, palette = "muted")
    g.despine(left=True)
    g = g.set_ylabels("survival probability")
    print(data[[feature, 'Survived']].groupby([feature]).agg(['mean', 'count']))
        
    #plt.hist(data['Fare'])
    #plt.scatter(data['Survived'], training_data['Age'])
    #plt.show()

def compare_float_to_label(data, feature):
    g = sns.distplot(data[feature], color="m", label="Skewness : %.2f"%(data[feature].skew()))
    g = g.legend(loc="best")
    
    g = sns.FacetGrid(data, col='Survived')
    g = g.map(sns.distplot, feature)
```

```python
#Cabin
compare_integer_to_label(ferature_eng(train), 'Cabin')
```
